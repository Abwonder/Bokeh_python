Q1. How can you create a Bokeh plot using Python code?

creating a Bokeh plot using python invovles the following processes:

1. Installation of Bokeh library

    *!pip install bokeh*
    
    
2. import the necessary modules needed for making plots

      *bokeh.io.output_notebook()*  ## Enable you to run on jupyter notebook
      
      
3. creat the plot title using

4. Create the actual plot using below code to render the x, y cordinates and other parameters like color, size etc.

<!-- See below for real example using dummy data -->

```python
## I already install Bokeh
#pip install bokeh

import bokeh.io
from bokeh.plotting import figure, output_file, show
bokeh.io.output_notebook()  ## Enable you to run on jupyter notebook

# create a new plot with a title
p = figure(title="My Bokeh Plot", x_axis_label="X-axis", y_axis_label="Y-axis")


# Create the actual plot using below code to render the x, y cordinates and other parameters like color, size etc
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=10, color="navy", alpha=0.5)


# specify where the output HTML file will be saved
output_file("my_plot.html")

# show the plot in a web browser
show(p)
```

### Question 2

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Glyph is a visual mark that represents data on a plot and the following are some examples of glyphs which include circles, squares, lines, bars, and patches. They are used to create visualizations that show how data is distributed or how it changes over time.

They can be added to Bokeh plot by calling them in functions used, check below example to explain this content.


```python
## libraries

import bokeh.io
from bokeh.plotting import figure, output_file, show
bokeh.io.output_notebook()  ## Enable you to run on jupyter notebook

### Creating the dummy dataset to be used
# create some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
```

```python
### first example for circle
## creat the title and axis labels
p = figure(title="Simple Circle Glyph", x_axis_label="x", y_axis_label="y")

### adding cicle glyph (creating)
p.circle(x,y,size=15,color="red", alpha=0.5)

## specify the output file and show the plot
output_file("circle_glyph.html")
show(p)
```

```python
### first example for line
## creat the title and axis labels
p = figure(title="Simple Line Glyph", x_axis_label="x", y_axis_label="y")

### adding cicle glyph (creating)
p.line(x,y,color="red", alpha=0.5)

## specify the output file and show the plot
output_file("Line_glyph.html")
show(p)
```

### Question 4

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Bokeh server is a service-base framework offered by bokeh whichh allows the creation for interactive plots that can be updated in real time.

It can be created by using python code and then run the application on a server, which allows users to interact with the application through a web browser.

To create a Bokeh server application, you start by defining your plot using Bokeh's plotting API. You can add widgets to the plot, such as sliders or buttons, which can be used to update the data or change the plot's appearance. You can also define callbacks that are triggered when the user interacts with the plot, allowing you to update the plot dynamically based on user input.

Once you have defined your plot and added any necessary widgets and callbacks, you can start the Bokeh server and run your application. The server handles the communication between the browser and the Python code, allowing the user to interact with the plot in real time.



```python
### Code example of the above question
from bokeh.layouts import column
from bokeh.models import Slider
from bokeh.plotting import figure, curdoc

# create a plot with a line and a slider
plot = figure(plot_width=400, plot_height=400)
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
line = plot.line(x, y)
slider = Slider(start=1, end=10, step=1, value=1)

# define a callback function that updates the line data
def update_data(attrname, old, new):
    line.data_source.data['y'] = [i*slider.value for i in x]

# attach the callback to the slider
slider.on_change('value', update_data)

# create a layout and add it to the current document
layout = column(slider, plot)
curdoc().add_root(layout)
```
